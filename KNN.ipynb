```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
data = datasets.load_wine(as_frame = True)
```

```python
X = data.data
Y = data.target
names = data.feature_names
print(names)
```

```python
df = pd.DataFrame(X, columns = names)
df['wine class']= data.target
df['wine class']=df['wine class'].replace(to_replace=[0,1,2],value=['class_0','class_1','class_2'])
```

```python
df
```

```python
sns.pairplot(data=df, hue= 'wine class', palette = 'Set2')
plt.show()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.3, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import math
math.sqrt(len(Y_test))
```

```python

knn = KNeighborsClassifier(n_neighbors = 7)  # Correct the typo: 'n_neighbours' should be 'n_neighbors'
knn.fit(X_train,Y_train)
pred = knn.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(Y_test,pred)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier # Make sure you import with the correct case

knn1 = KNeighborsClassifier(n_neighbors = 7, metric = 'euclidean') # Change 'metrics' to 'metric'
knn1.fit(X_train,Y_train)
pred2 = knn1.predict(X_test)
```

```python
metrics.accuracy_score(Y_test,pred2)
```
